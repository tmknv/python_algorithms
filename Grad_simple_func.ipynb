```python
x0=5
a=0.01
epsilon=0.001
num_iterations=10000
```

```python
def new_weight(x):
 return 4*(x**3)-10
```

```python
# import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**4-10*x+3
x_val=[x for x in np.arange(-5,5,0.1)]
y_val=[f(x) for x in np.arange(-5,5,0.1)]

plt.figure(figsize=(8,8))
plt.axvline(x=0, c='black')
plt.axhline(y=0, c='black')
plt.plot(x_val,y_val)
```

```python
def grad_calculation(x0,a,epsilon,num_iterations):
    x_curr=x0
    for _ in range(num_iterations):
        new_x=x_curr-a*new_weight(x_curr)

        if abs(new_x-x_curr) < epsilon:
            return x_curr
        x_curr=new_x
    return x_curr
```

```python
x_res=grad_calculation(x0,a,epsilon,num_iterations)
x_res
```

```python
plt.figure(figsize=(8,8))
plt.axvline(x=0, c='black')
plt.axhline(y=0, c='black')
plt.plot(x_val,y_val)
plt.scatter(x_res,f(x_res),marker='o',c='red' )
```
